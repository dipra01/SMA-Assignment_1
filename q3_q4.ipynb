```python
from networkx.algorithms.community.modularity_max import greedy_modularity_communities
import time
import networkx as nx
import networkx.algorithms.community as nx_comm

graph1 = nx.read_gml('karate.gml', label = 'id')
graph2 = nx.read_gml('dolphins.gml')
graph3 = nx.read_weighted_edgelist('jazz.net')

print("==================Modularity Maximization Algorithm===================")
print("------------------For Karate Club Network-------------------------")
# Graph1
seconds_1 = time.time()

# find communities in the graph
comm1 = greedy_modularity_communities(graph1)

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

# find the nodes forming the communities
node_groups1 = (list(sorted(c) for c in comm1))

print("Clusters : ")
print(node_groups1)   
print("Number of Clusters : ")
print(len(node_groups1))  
print("Modularity : " + str(nx_comm.modularity(graph1, node_groups1)))
print("\n===========================================================================================\n")

print("------------------For Dolphins Social Network--------------------")

# Graph2
seconds_1 = time.time()

# find communities in the graph
comm2 = greedy_modularity_communities(graph2)

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

# find the nodes forming the communities
node_groups2 = (list(sorted(c) for c in comm2))

print("Clusters : ")
print(node_groups2)   
print("Number of Clusters : ")
print(len(node_groups2))  
print("Modularity : " + str(nx_comm.modularity(graph2, node_groups2)))
print("\n===========================================================================================\n")



print("--------------------For Jazz Musicians Network----------------------")

# Graph3
seconds_1 = time.time()

# find communities in the graph
comm3 = greedy_modularity_communities(graph3)

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

# find the nodes forming the communities
node_groups3 = (list(sorted(c) for c in comm3))

print("Clusters : ")
print(node_groups3)   
print("Number of Clusters : ")
print(len(node_groups3))  
print("Modularity : " + str(nx_comm.modularity(graph3, node_groups3)))
```

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
from numpy import random
import time
import networkx as nx
import networkx.algorithms.community as nx_comm

graph1 = nx.read_gml('karate.gml', label = 'id')
graph2 = nx.read_gml('dolphins.gml')
graph3 = nx.read_weighted_edgelist('jazz.net')

print("==================Spectral Clustering Algorithm===================")

# Graph1
print("----------------------For Karate Club Network---------------------")
adj_matrix1 = nx.to_numpy_matrix(graph1)

seconds_1 = time.time()

sc1 = SpectralClustering(n_clusters=4, affinity='precomputed', n_init=100)

# find communities in the graph
sc1.fit(adj_matrix1)

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

# find the nodes forming the communities
labels = sc1.labels_

lst = [[] for i in range(4)]

i = 1
for x in labels:
    if(x == 0):
        lst[0].append(i)
        i += 1
    elif(x == 1):
        lst[1].append(i)
        i += 1
    elif(x == 2):
        lst[2].append(i)
        i += 1
    elif(x == 3):
        lst[3].append(i)
        i += 1
        

print("Clusters : ")
print(lst)
print("Number of Clusters : ")
print(len(lst)) 
print("Modularity : " + str(nx_comm.modularity(graph1, lst)))
print("=====================================================================================")


# Graph2
print("-------------------------For Dolphins Social Network------------------------")

adj_matrix2 = nx.to_numpy_matrix(graph2)

# convert graph to a list
lst_graph2 = list(graph2)

seconds_1 = time.time()

sc2 = SpectralClustering(n_clusters=4, affinity='precomputed', n_init=100)

# find communities in the graph
sc2.fit(adj_matrix2)

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

# find the nodes forming the communities
labels = sc2.labels_

lst = [[] for i in range(4)]

i = 0
for x in labels:
    if(x == 0):
        lst[0].append(lst_graph2[i])
        i += 1
    elif(x == 1):
        lst[1].append(lst_graph2[i])
        i += 1
    elif(x == 2):
        lst[2].append(lst_graph2[i])
        i += 1
    elif(x == 3):
        lst[3].append(lst_graph2[i])
        i += 1
        

print("Clusters : ")
print(lst)
print("Number of Clusters : ")
print(len(lst)) 
print("Modularity : " + str(nx_comm.modularity(graph2, lst)))
print("=====================================================================================")



# Graph3
print("--------------------For Jazz Musicians Network---------------------")


adj_matrix3 = nx.to_numpy_matrix(graph3)

seconds_1 = time.time()

sc3 = SpectralClustering(n_clusters=4, affinity='precomputed', n_init=100)

# find communities in the graph
sc3.fit(adj_matrix3)

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

# find the nodes forming the communities
labels = sc3.labels_

lst = [[] for i in range(4)]

i = 1
for x in labels:
    if(x == 0):
        lst[0].append(str(i))
        i += 1
    elif(x == 1):
        lst[1].append(str(i))
        i += 1
    elif(x == 2):
        lst[2].append(str(i))
        i += 1
    elif(x == 3):
        lst[3].append(str(i))
        i += 1
        

print("Clusters : ")
print(lst)
print("Number of Clusters : ")
print(len(lst)) 
print("Modularity : " + str(nx_comm.modularity(graph3, lst)))
print("=====================================================================================")
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import networkx.algorithms.community as nx_comm
import time


def edge_to_remove(graph):
    Graph_dict = nx.edge_betweenness_centrality(graph)
    edge = ()

  # extract the edge with highest edge betweenness centrality score
    for key, value in sorted(Graph_dict.items(), key=lambda item: item[1], reverse = True):
        edge = key
        break
        
    return edge


def girvan_newman(graph):
    # find number of connected components
    cc = nx.connected_components(graph)
    cc_count = nx.number_connected_components(graph)

    while(cc_count == 1):
        graph.remove_edge(edge_to_remove(graph)[0], edge_to_remove(graph)[1])
        cc = nx.connected_components(graph)
        cc_count = nx.number_connected_components(graph)

    return cc

graph1 = nx.read_gml('karate.gml', label = 'id')
graph2 = nx.read_gml('dolphins.gml')
graph3 = nx.read_weighted_edgelist('jazz.net')

print("==================Girvan Newman Algorithm===================")

print("------------------For Karate Club Network-------------------")
# Graph1
seconds_1 = time.time()

# find communities in the graph
comm1 = girvan_newman(graph1.copy())

# find the nodes forming the communities
node_groups1 = []

for i in comm1:
    node_groups1.append(list(i))

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

print("Clusters : ")
print(node_groups1)   
print("Number of Clusters : ")
print(len(node_groups1))   
print("Modularity : " + str(nx_comm.modularity(graph1, node_groups1)))
print("===========================================================================================")

# Graph2
print("-------------------For Dolphins Social Network----------------")
seconds_1 = time.time()

# find communities in the graph
comm2 = girvan_newman(graph2.copy())

# find the nodes forming the communities
node_groups2 = []

for i in comm2:
    node_groups2.append(list(i))

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

print("Clusters : ")
print(node_groups2)    
print("Number of Clusters : ")
print(len(node_groups2))  
print("Modularity : " + str(nx_comm.modularity(graph2, node_groups2)))
print("\n===========================================================================================\n")

# Graph3
print("--------------------For Jazz Musicians Network---------------")
seconds_1 = time.time()

# find communities in the graph
comm3 = girvan_newman(graph3.copy())

# find the nodes forming the communities
node_groups3 = []

for i in comm3:
    node_groups3.append(list(i))

seconds_2 = time.time()
print("Time taken : ", seconds_2 - seconds_1)

print("Clusters : ")
print(node_groups3)  
print("Number of Clusters : ")
print(len(node_groups3))  
print("Modularity : " + str(nx_comm.modularity(graph3, node_groups3)))
```
